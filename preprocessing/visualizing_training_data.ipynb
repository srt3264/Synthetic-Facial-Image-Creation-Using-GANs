```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#count the number of files in the directory and subdirectories 

#count the number of files in the directory and subdirectories
import os
def count_files(directory):
    count = 0
    for root, dirs, files in os.walk(directory):
        count += len(files)
    return count

happy_train = count_files('/workspaces/Team-8-Machine-Learning-1/data/train/happy')
neutral_train = count_files('/workspaces/Team-8-Machine-Learning-1/data/train/neutral')
sad_train = count_files('/workspaces/Team-8-Machine-Learning-1/data/train/sad')
angry_train = count_files('/workspaces/Team-8-Machine-Learning-1/data/train/angry')
fear_train = count_files('/workspaces/Team-8-Machine-Learning-1/data/train/fear')
disgust_train = count_files('/workspaces/Team-8-Machine-Learning-1/data/train/disgust')
surprise_train = count_files('/workspaces/Team-8-Machine-Learning-1/data/train/surprise')

```

```python
# create a histogram of the number of files in each directory
#make plot have retina display
%config InlineBackend.figure_format = 'retina'
plt.bar(['happy', 'neutral', 'sad', 'angry', 'fear', 'disgust', 'surprise'], [happy_train, neutral_train, sad_train, angry_train, fear_train, disgust_train, surprise_train], color=('purple','pink', 'red', 'hotpink', 'deeppink', 'palevioletred', 'mediumvioletred'))
plt.title('Number of Images for Each Emotion in the Training Set')
plt.xlabel('Emotion')
plt.ylabel('Count of Images')
plt.show()
```
