```python
# path = '"/workspaces/Team-8-Machine-Learning/data"'

import os
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt


path_train_1 = '/Users/ahmedibrahim/Desktop/Mids/Spring23/machine_learning/ML_Project2/Team-8-Machine-Learning/data_processed/filtered/train/angry/'


# Count the number of images.
num_images = 0
for subdir, dirs, files in os.walk(path_train_1):
    for filename in files:
        if filename.endswith(".jpg") or filename.endswith(".png"):
            num_images += 1

print(num_images)

#create a dictionary that stores image names and their corresponding SNR values

img_angry = {}

snr_angry = []

img_names = []

for subdir, dirs, files in os.walk(path_train_1):
    for filename in files:
        if filename.endswith(".jpg") or filename.endswith(".png"):
            # Open the image.
            img = Image.open(os.path.join(subdir, filename))

            #store the name of the image and only get the file name

            img_name = os.path.join(subdir, filename)

            #remove the path from the image name

            img_name = img_name.replace(path_train_1, '')

            # Convert the image to grayscale.
            img = img.convert('L')

            # Get the image data.
            # data = img.getdata()

            # Convert the image data to a numpy array.
            data = np.array(img)

        

            #calculate the signal (room mean pixel value)

            signal = np.mean(data)




            # Calculate the noise (standard deviation of pixel values)

            noise = np.std(data)


            # Calculate SNR in decibels (dB)

            snr = 20 * np.log10(signal / noise)

            
            snr_angry.append(snr)

            img_names.append(img_name)

            #store the image name and SNR value in the dictionary

            img_angry[img_name] = snr



    print(snr_angry)

    print(len(snr_angry))

    # print(img_names)

    print(img_angry)

```

```python
#classify images based on SNR values if SNR > 3 then the image is clear and if SNR < 3 then the image is noisy

#store the image names in a list

img_names_angry_clear = []

img_names_angry_noisy = []

for key, value in img_angry.items():

    if value > 1:

        img_names_angry_clear.append(key)

    if value < 1:

        img_names_angry_noisy.append(key)

#store the corresponding images into a folder

#clear images

for i in range(len(img_names_angry_clear)):

    img = Image.open(os.path.join(path_train_1, img_names_angry_clear[i]))

    img.save('/Users/ahmedibrahim/Desktop/Mids/Spring23/machine_learning/ML_Project2/Team-8-Machine-Learning/data_processed/Noise_removed/noise_train/angry_clear/' + img_names_angry_clear[i])

#noisy images

for i in range(len(img_names_angry_noisy)):

    img = Image.open(os.path.join(path_train_1, img_names_angry_noisy[i]))

    img.save('/Users/ahmedibrahim/Desktop/Mids/Spring23/machine_learning/ML_Project2/Team-8-Machine-Learning/data_processed/Noise_removed/noise_train/angry_noisy/' + img_names_angry_noisy[i])
```
